### Step2. Setup iterative coupling solver

```python
BEMCases=[BEM_Case1,BEM_Case2,BEM_Case3,BEM_Case4,BEM_Case5,BEM_Case6,BEM_Case7,BEM_Case8,BEM_Case9] #BEM_Case1,BEM_Case2 are coupled
Intersections=[]
#Intersection 1
#Domain 0 <-> Domain 1
#BD 1 <-> BD 3
DomainID,BoundaryID=0,1
DomainID_connect,BoundaryID_connect=1,2
Intersections.append([DomainID,DomainID_connect,BoundaryID,BoundaryID_connect])

DomainID,BoundaryID=0,3
DomainID_connect,BoundaryID_connect=2,2
Intersections.append([DomainID,DomainID_connect,BoundaryID,BoundaryID_connect])

DomainID,BoundaryID=0,5
DomainID_connect,BoundaryID_connect=3,2
Intersections.append([DomainID,DomainID_connect,BoundaryID,BoundaryID_connect])

DomainID,BoundaryID=0,7
DomainID_connect,BoundaryID_connect=4,2
Intersections.append([DomainID,DomainID_connect,BoundaryID,BoundaryID_connect])

DomainID,BoundaryID=0,11
DomainID_connect,BoundaryID_connect=5,0
Intersections.append([DomainID,DomainID_connect,BoundaryID,BoundaryID_connect])

DomainID,BoundaryID=0,13
DomainID_connect,BoundaryID_connect=6,0
Intersections.append([DomainID,DomainID_connect,BoundaryID,BoundaryID_connect])

DomainID,BoundaryID=0,15
DomainID_connect,BoundaryID_connect=7,0
Intersections.append([DomainID,DomainID_connect,BoundaryID,BoundaryID_connect])

DomainID,BoundaryID=0,17
DomainID_connect,BoundaryID_connect=8,0
Intersections.append([DomainID,DomainID_connect,BoundaryID,BoundaryID_connect])

BEMSolver=BEMpy.DDM_Solver(BEMobj=BEMCases,Intersection=Intersections)
```

### Step3. Parallel Robin-Robin Algorithm

```python
BEMSolver=BEMpy.DDM_Solver(BEMobj=BEMCases,Intersection=Intersections)
BEMSolver.Solve_Iter(Method="P-RR",TOL=1e-5,alpha=1.0,opt=1)
```

### Step3. Parallel Neumann-Neumann Algorithm

```python
BEMSolver=BEMpy.DDM_Solver(BEMobj=BEMCases,Intersection=Intersections)
BEMSolver.Solve_Iter(Method="P-NN",TOL=1e-5,max_iters=200,alpha=0.1,opt=1)
```

```python
BEMSolver=BEMpy.DDM_Solver(BEMobj=BEMCases,Intersection=Intersections)
BEMSolver.Solve_Iter(Method="S-DN",TOL=1e-5,alpha=0.1,opt=1)
```
